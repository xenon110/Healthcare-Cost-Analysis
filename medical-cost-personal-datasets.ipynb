```python
non_smoker_df = datas[datas.smoker == 'no']
```

```python
def estimate_charges(age,w,b):
    return w * age + b
```

```python
w=50
b=100
```

```python
ages = non_smoker_df.age
print(ages)
```

```python
estimated_charges = estimate_charges(ages,w,b)
estimated_charges
```

```python
non_smoker_df.charges
```

```python
predication = estimated_charges
predication
```

```python
def rmse(target,predicatioin):
    return np.sqrt(np.mean(np.square(target-predication)))
```

```python
w = 50
b = 100
target = non_smoker_df['charges']
predicated = estimate_charges(non_smoker_df.age,w,b)
```

```python
rmse(target,predicated)
```

```python
# Now we will us the module to calculate this
from sklearn.linear_model import LinearRegression
```

```python
model =LinearRegression()
```

```python
input = non_smoker_df[['age']]
target = non_smoker_df.charges
print('input.shape:',input.shape)
print('target.shape:',target.shape)
```

```python
model.fit(input,target)
```

```python
predications = model.predict(input)
predications
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(target, predications))
print("RMSE:", rmse)
```
